# Benchmark YOLOv8 vs DETR

Run both models, visualize FPS and consistency.

```python
import json, pandas as pd, subprocess, sys
from pathlib import Path
video = 'crowd.mp4'  # change if needed
out_yolo = 'out_yolo.mp4'
out_detr = 'out_detr.mp4'

# Run YOLO
subprocess.run([sys.executable, '-m', 'src.infer', '--model','yolo','--input',video,'--output',out_yolo,'--device','auto','--conf','0.30','--resize','1280'], check=False)
# Run DETR
subprocess.run([sys.executable, '-m', 'src.infer', '--model','detr','--input',video,'--output',out_detr,'--device','auto','--conf','0.30','--resize','1280'], check=False)

def load_json(p):
    try:
        with open(p,'r') as f: return json.load(f)
    except: return {}

m_y = load_json('out_yolo_metrics_yolo.json')
m_d = load_json('out_detr_metrics_detr.json')
pd.DataFrame([m_y,m_d], index=['YOLOv8n','DETR'])
```
